```python
import os
os.chdir(r"../../..")
print(os.getcwd())
```

```python
#Imports


from src.training.dataset_processing_pipeline.dataset_generator import process_dataset
from src.training.training_pipeline.training_utils import get_datasets, train_model
from src.utils.model_builder import build_model
```

```python
#Parameters

dataset_limit = 2
batch_size = 2048
validation_split = 0.05

input_shape = (8, 8, 12)
blocks = 6
filters = 64
se_ratio = 8

epochs=100
initial_epoch=0
learning_rate=1e-3
policy_loss_weight=1.0
value_loss_weight=1.6
use_mixed_precision=True
run_name=r"test1"
base_log_dir="../../../logs"
base_model_dir="../../../nets"
patience_stop=5
patience_lr=3
lr_factor=0.1
save_best_only=True
```

```python
#Dataset writing

process_dataset(limit=dataset_limit)
```

```python
#Dataset spliting

training_dataset, validation_dataset, train_steps, validation_steps = get_datasets(tfrecord_dir=r"data/tfrecords_gzip", batch_size=batch_size, validation_split=validation_split)
```

```python
#Model building

model = build_model(input_shape=input_shape, blocks=blocks, filters=filters, se_ratio=se_ratio)
```

```python
#Starting tensorboard (has to be installed outside the venv)

%load_ext tensorboard

%tensorboard --logdir logs --host 0.0.0.0 --port 6006
```

```python
#Training process

history = train_model(model, training_dataset, validation_dataset, train_steps=train_steps, validation_steps=validation_steps, epochs=epochs, initial_epoch=initial_epoch, learning_rate=learning_rate, policy_loss_weight=policy_loss_weight, value_loss_weight=value_loss_weight, use_mixed_precision=use_mixed_precision, run_name=run_name, base_log_dir=base_log_dir, base_model_dir=base_model_dir, patience_stop=patience_stop, patience_lr=patience_lr, lr_factor=lr_factor, save_best_only=save_best_only)

```

```python
#Saving

model.save(f"nets/{run_name}.keras")
```
